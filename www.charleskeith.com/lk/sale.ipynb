```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import lxml
from xlwt import *
```

```python
URL = 'https://www.charleskeith.com/lk/sale'
page = requests.get(URL)
```

```python
soup = BeautifulSoup(page.content, 'lxml')

data = pd.DataFrame(columns=['Product ID', 'Product Name', 'Current Price', 'Previous Price', 'Discount'])
cards = soup.find_all('div', class_='col-6 col-xl-4 product_tile col-lg-4 js-product_tile js-product_list_element')

for card in cards:
  discount = card.find('div', class_='price').text
  product_no = card.find('div', class_='pdp-link product_tile-name').a['href'].split('/')[-1]
  product_name = card.find('div', class_='product-tile ').a.text.strip()
  cur_price = card.find('span', class_='price-wrap').text.strip()
  prev_price = card.find('span', class_= 'value').text.strip()

  data_dict = {
      'Product ID': product_no,
      'Product Name': product_name,
      'Current Price': cur_price,
      'Previous Price': prev_price,
      'Discount': discount
  }

  df = pd.DataFrame(data_dict, index=[0])
  data = data.append(df, ignore_index=True)


  print("Discount: {discount} | Product ID: {product_no} | Product Name: {product_name} | Previous Price: {prev_price} | Price: {cur_price}")
```
