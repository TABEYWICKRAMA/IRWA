Question 01 - Implement a program to test if a given page is found or not on the server.

#Question 04 - How to print all the header tags from en.wikipedia.org/wiki/Main_Page. 
####BeautifulSoup(html, "html.parser")
####html.parser - get all the html data
####lxml - get all html and xml data
####BeutifulSoup(html) default one is html.parser

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
html = urlopen('https://en.wikipedia.org/wiki/Main_Page')
bs = BeautifulSoup(html)  #to scrap data from source code
titles = bs.find_all(['h1','h2','h3'])         #"find_all" --->> this will print all the header tags(h1, h2 and h3).     #"find" will print first header tag only(h1). 
print('List the first header tag: ', *titles, sep='\n\n')    # we use "*" to get headers in separate lists. if we do not use "*" then all the headers will print in one list.
```

Question 05 - How to display all the image links from https://www.ibm.com/lk-en. 

```python
#method 01
import requests 
from bs4 import BeautifulSoup 
    
def getdata(url): 
    r = requests.get(url) 
    return r.text 
    
htmldata = getdata("https://www.geeksforgeeks.org/") 
soup = BeautifulSoup(htmldata, 'html.parser') 
for item in soup.find_all('img'):
    print(item['src'])
```

Question 7 - Python implementation to display section that contain a specified string in a given web 
page.

```python
#string should come as the regular expressions.

import requests
from bs4 import BeautifulSoup
import re

url = "https://en.wikipedia.org/wiki/Main_Page"
req = requests.get(url)
bs = BeautifulSoup(req.text, 'lxml')
str1 = bs.find_all(string=re.compile('Hurricane Julia'))
for txt in str1:
  print(txt)
```

```python
import requests
from bs4 import BeautifulSoup
url = "https://www.sliit.lk/"
req = requests.get(url)
bs = BeautifulSoup(req.text,'lxml')
print(bs.title)   #this will print title of this website
print(bs.title.text) #this will print without the tags content only

print(bs.title.parent)  #this will show all the content above the title tag
```
